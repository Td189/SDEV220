```python
'''
Assign the value 5 to the variable guess_me. Use a for loop to iterate a variable called number over range(10). 
If number is less than guess_me, print 'too low'. If it equals guess_me, print found it! and then break out of the for loop.
If number is greater than guess_me, print 'oops' and then exit the loop.
'''
```

```python
list = [3,2,1,0]
for n in list:
    print(n)
guess_me = 7
number = 1
running = True
while running:
    if number < guess_me:
        print("too low!\n")
        number += 1
    elif number > guess_me:
        print("oops!\n")
        number -= 1
    elif number == guess_me:
        print("found it!\n")
        running = False
        break
guess_me = 5
for number in range(10):
    if number < guess_me:
        print("too low!\n")
    if number == guess_me:
        print("found it!\n")
        break
    if number > guess_me:
        print("oops\n")
        break
        
```
