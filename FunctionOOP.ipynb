```python
import random as r
class Solution:
    def firstSearch(self, arr, k):
        # Code Here
      left = 0
      right = len(arr) - 1
      result = -1
    
      while left <= right:
        mid = (left + right) // 2
    
        if arr[mid] == k:
            result = mid
            right = mid - 1
    
        if arr[mid] < k:
          left = mid + 1
        else:
          right = mid - 1
    
      return result
        
    def countingSort(self, arr, place):
        size = len(arr)
        output = [0] * size
        count = [0] * 10

        for i in range(size):
            index = arr[i] // place
            count[index % 10] += 1

        for i in range(1, 10):
            count[i] += count[i - 1]

        i = size - 1
        while i >= 0:
            index = arr[i] // place
            output[count[index % 10] - 1] = arr[i]
            count[index % 10] -= 1
            i -= 1

        for i in range(size):
            arr[i] = output[i]

    def radixSort(self, arr):
        max_element = max(arr)

        place = 1
        while max_element // place > 0:
            self.countingSort(arr, place)
            place *= 10
ls = [r.randint(1,10) for i in range(10)]
s = Solution()
x = input("Enter an integer between 1 and 10 inclusive")
print(ls)
print(s.firstSearch(ls,5))
print(s.radixSort(ls))

      
```
