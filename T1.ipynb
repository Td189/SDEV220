```python
import random as r
running = True
while running:
    secret = r.randint(1,10)
    guess = r.randint(1,10)
    print(guess)
    if(guess == secret):
        print("just right!\n")
    elif guess < secret :
        print("too low!\n")
    elif guess > secret : 
        print("too high!\n")
    ans = int(input("Try again ?\n1 : Yes\n0 : No\n"))
    running = True if ans == 1 else False
print("Thank you for using the program!\n")
    
```
